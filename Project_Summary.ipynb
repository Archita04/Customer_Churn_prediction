**Min Max scaler** - Min Max scaler is sclaing technique that scales all the data features in the range [0, 1].  * **Data Balancing** - 
 
**SMOTE()** - It is an oversampling technique aims to balance class distribution by randomly increasing minority class examples by replicating them.Our data is seems balanced but to get more accuracy used SMOTE and any model without balancing would be a dump model..For balancing the data i used SMOTE() oversampling technique...
